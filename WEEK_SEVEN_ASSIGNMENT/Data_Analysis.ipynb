```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = "sample_dataset.csv"  # Ensure the CSV file is in the same directory
df = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
head_data = df.head()
head_data
```

```python
# Get information about the dataset
info_data = df.info()
```

```python
# Summary statistics
summary_stats = df.describe()
summary_stats
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
# Grouping data by Department and calculating mean Salary
mean_salary_by_department = df.groupby("Department")["Salary"].mean()
mean_salary_by_department
```

```python
# 1. Histogram for Age Distribution
plt.figure(figsize=(8, 5))
df['Age'].hist(bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(False)
plt.show()
```

```python
# 2. Bar chart for Mean Salary by Department
plt.figure(figsize=(8, 5))
mean_salary_by_department.plot(kind='bar', color='coral', edgecolor='black')
plt.title('Mean Salary by Department', fontsize=14)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Mean Salary', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 3. Scatter plot of Salary vs. Years of Experience
plt.figure(figsize=(8, 5))
plt.scatter(df['Years_of_Experience'], df['Salary'], color='green', alpha=0.7)
plt.title('Salary vs. Years of Experience', fontsize=14)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Salary', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
